```python
nan_percentage = df.isnull().sum() * 100 / len(df)  # hna 7sbt qema kol col hwa fe null be nsbt ad eh 
print(nan_percentage)
```

```python
df=df.drop(['WindGustSpd','PoorWeather','PRCP','DR','SPD','SND','FT','FB','FTI','ITH','PGT','RHN','RVG','WTE','TSHDSBRSGF','SD3','RHX'],axis=1)   
df
```

```python
df['Snowfall']=df['Snowfall'].replace("#VALUE!",0)
df['Snowfall']=pd.to_numeric(df['Snowfall'])
df['Snowfall']=df['Snowfall'].fillna(df['Snowfall'].mean())
```

```python
df['SNF']=df['SNF'].replace('T',0)
df['SNF']=pd.to_numeric(df['SNF'])
```

```python
df['MAX']=df['MAX'].fillna(df['MAX'].mean())
df['MIN']=df['MIN'].fillna(df['MIN'].mean())
df['MEA']=df['MEA'].fillna(df['MEA'].mean())
df['SNF']=df['SNF'].fillna(df['SNF'].mean())
```

```python
df.isna().sum()
```

```python
df['Date']=pd.to_datetime(df['Date'])
df['year']=df['Date'].dt.year
df['month']=df['Date'].dt.month
df['day']=df['Date'].dt.day
```

```python
df.drop(['Date'],axis=1)
```

```python
df['Precip']=df['Precip'].replace('T',0)
df['Precip']=pd.to_numeric(df['Precip'], errors='coerce')
df['Precip']=df['Precip'].fillna(0)
```

```python
plt.figure(figsize=(12,8))
numeriac_data=df.select_dtypes(include=('float64','int64'))
sns.heatmap(numeriac_data.corr(),annot=True)
```

# preprocessing

```python
from sklearn.model_selection import train_test_split
```

```python
x= df[['MinTemp', 'Precip','year','month', 'day']]
y=df['MaxTemp']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Model

```python
from sklearn.ensemble import RandomForestRegressor # r2 score 80
from sklearn.linear_model import LinearRegression # r2 score 78
from sklearn.metrics import mean_squared_error,r2_score
```

```python
RF=RandomForestRegressor()
```

```python
RF.fit(x_train,y_train)
```

```python
y_pred=RF.predict(x_test)
```

```python
y_test
```

```python
y_pred
```

```python
mse=mean_squared_error(y_test,y_pred)
print("Mse score ",mse)
```

```python
r2=r2_score(y_test,y_pred)
print("r2 score ",r2)
```
